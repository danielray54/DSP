```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import math as maths
from scipy.interpolate import interp1d 
```

```python
df = pd.read_csv("Data/PPG_01.csv")

plt.title("Heart Rate Signal") 
df.ppg.plot(figsize=(20, 5)) 
plt.ylabel('Amplitude')
plt.xlabel('Frequency')
plt.show()
```

```python
print('Number of samples', len(df.ppg))
hz = 50
print(hz, 'Hz')
print('Timestep (ms)', 1 / hz * 1000)
print("Length of Signal (s): ", len(df.ppg) * (1 / hz) )
```

```python
#--Moving Average--
#One-sided window size, as proportion of the sampling frequency
hrw = 0.75 
#Calculate moving average
moving_avg = df['ppg'].rolling(int(hrw*hz)).mean()
moving_avg
```

```python
#Compute avg
avg_hr = (np.mean(df.ppg))
#add avg for NaN at beginning
mov_avg = [avg_hr if maths.isnan(x) else x for x in moving_avg]
#raise the average by 20% to prevent the secondary heart contraction from interfering
#mov_avg = [x*1.2 for x in mov_avg] 
#Append the moving average to the dataframe
df['heart_rollingmean'] = mov_avg 
df['heart_rollingmean']
```

```python
#--Mark regions of interest--
window = []
peaklist = []
listpos = 0

#for each datapoint in df
for x in df.ppg:
    #get local avg
    rolling_mean = df.heart_rollingmean[listpos]
    #if datapoint is less than local avg and there are no points in ROI
    if(x < rolling_mean) and (len(window) < 1):
        #inc counter (do nothing)
        listpos += 1
    #if datapoint is greater than local avg
    elif(x > rolling_mean):
        #Mark it as ROI
        window.append(x)
        #inc counter
        listpos += 1
    #if x is less than local avg, i.e. region of intrest has passed
    else:
        #determine the highest point in window (ROI)
        maximum_val = max(window)
        #Position of the point on the X-axis
        #listpos is at end of ROI so minus length of window to go back to start of ROI and add the index value to get pos on x axis
        beat_pos = listpos - len(window) +(window.index(maximum_val))
        #add to list
        peaklist.append(beat_pos)
        #clear ROI
        window = []
        #continue to next datapoint
        listpos += 1
```

```python
def merge(list1, list2): 
      
    merged_list = [(list1[i], list2[i]) for i in range(0, len(list1))] 
    return merged_list 
```

```python
ybeat = [df.ppg[x] for x in peaklist]
print(merge(peaklist, ybeat))
```

```python
plt.figure(figsize=(20,5))
plt.title("Detected peaks in signal")
plt.plot(df.ppg, alpha=0.5, color='blue') #Plot semi-transparent HR
plt.plot(df['heart_rollingmean'], color ='green') #Plot moving average
plt.scatter(peaklist, ybeat, color='red') #Plot detected peaks
plt.show()
```

```python
#list of r-r intervals
RR_list=[]
i = 0

while (i < (len(peaklist)-1)):
    #Calculate distance between beats in # of samples
    RR_interval = (peaklist[i+1]-peaklist[i])
    #convert to ms
    ms_dist = ((RR_interval/hz)*1000)
    #add to list of rr intervals
    RR_list.append(ms_dist)
    #continue
    i+=1

bpm = 60000/np.mean(RR_list)
print ("Average Heart Beat is: %.02f" %bpm) 
```

```python
RR_list
```

```python
plt.figure(figsize=(20,5))
plt.plot([60000/x  for x in RR_list], color ='green')
plt.hlines(60000/np.mean(RR_list),0, len([60000/x  for x in RR_list])-1, color='r')
plt.title('Moving Avg HR')
plt.ylabel('HR')
plt.xlabel('Seconds')
plt.show()
```

```python
RR_diff = []
RR_sqdiff = []
RR_list = RR_list
i = 1

while (i < (len(RR_list)-1)):
    RR_diff.append(abs(RR_list[i] - RR_list[i+1]))
    RR_sqdiff.append(maths.pow(RR_list[i] - RR_list[i+1], 2))
    i += 1
    
print(RR_diff, RR_sqdiff)
```

```python
ibi = np.mean(RR_list)

sdnn = np.std(RR_list)

sdsd = np.std(RR_diff)

rmssd = np.sqrt(np.mean(RR_sqdiff))
 
pnn20 = float(len([x for x in RR_diff if (x>20)]))/ float(len(RR_diff))

pnn50 = float(len([x for x in RR_diff if (x>50)]))/ float(len(RR_diff))

print("IBI: ", ibi)
print("SDNN: ", sdnn)
print("SDSD: ", sdsd)
print("RMSSD: ", rmssd)
print("PNN20: ", pnn20)
print("PNN50: ", pnn50)
```

```python
plt.figure(figsize=(20,5))
plt.plot( RR_list, color ='green')
plt.title('RR Intervals')

plt.show()
```

```python
#Remove the first entry, because first interval is assigned to the second beat.
RR_x = peaklist[1:]
#Y-values are equal to interval lengths
RR_y = RR_list 
#Create evenly spaced timeline starting at the second peak, its endpoint and length equal to position of last peak
RR_x_new = np.linspace(RR_x[0], RR_x[-1], RR_x[-1])
#Interpolate the signal with cubic spline interpolation
f = interp1d(RR_x, RR_y, kind='cubic') 
```

```python

plt.figure(figsize=(20,5))
plt.title("Original and Interpolated Signal")
plt.plot(RR_x, RR_y, label="Original", color='blue')
plt.plot(RR_x_new, f(RR_x_new), label="Interpolated", color='red')
plt.legend()
plt.show()
```

```python
n = len(df.ppg) #Length of the signal
frq = np.fft.fftfreq(len(df.ppg), d=((1/hz))) #divide the bins into frequency categories
frq = frq[range(int(n/2))] #Get single side of the frequency range
```

```python

#Do FFT
Y = np.fft.fft(f(RR_x_new))/n #Calculate FFT
Y = Y[range(int(n/2))] #Return one side of the FFT

#Plot
plt.figure(figsize=(20,5))
plt.title("Frequency Spectrum of Heart Rate Variability")
plt.xlim(0,0.6) #Limit X axis to frequencies of interest (0-0.6Hz for visibility, we are interested in 0.04-0.5)
plt.ylim(0, 50) #Limit Y axis for visibility
plt.plot(frq, abs(Y)) #Plot it
plt.xlabel("Frequencies in Hz")
plt.show()
```

```python
#Slice frequency spectrum where x is between 0.04 and 0.15Hz (LF), and use NumPy's trapezoidal integration function to find the area
lf = np.trapz(abs(Y[(frq>=0.04) & (frq<=0.15)])) 
print("LF:", lf)
#Do the same for 0.16-0.5Hz (HF)
hf = np.trapz(abs(Y[(frq>=0.16) & (frq<=0.5)])) 
print("HF:", hf)
```

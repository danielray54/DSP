## Signal Statistics and Noise

- Signal - How one parameter relates to another
    - if both parameters are continuous variables the signal is  <b>continuous </b>
    - when a continuous signal is passed through an analog to digital converter the two parameters become quantised, making the signal a <b> discrete signal </b>
        - where the y-axis is the dependent variable / range and the x-axis the independent variable / domain
        - the x-axis label is commonly refered to as sample number
        - the parameter of the y axis is a function of the parameter of the y-axis
        - N - Total number of samples (0 to N-1)

### Signal Mean and Standard Deviation

The mean of a signal is computed using:
$$\mu = \frac{1}{N}\sum_{i=0}^{N-1}x_{i}$$
Can be refered to as the DC component of the signal 

The deviation of each point is computed using:
$$\left | x_{i}-\mu \right |$$

The standard Deviation is computed using:
$$\sigma = \sqrt{\frac{1}{N-1}\sum_{i=0}^{N-1}(x_{i}-\mu)^{^{2}}}$$
- as power $P$ is directly proportional to voltage squared $V^{2}$
- measures the ac portion of the signal

The variance is computed using:
$$\sigma{^{2}} = \frac{1}{N-1}\sum_{i=0}^{N-1}(x_{i}-\mu)^{^{2}}$$
- measures the power of fluctuation

#### Computing the Mean

```python
#mean of the signal
sample_signal_mean = np.mean(data.InputSignal_1kHz_15kHz)
sample_signal_mean
```

```python
_mean = 0.0
def calc_signal_mean(sig_src):
    global _mean
    #for each item in the list, add item to total
    for item in range(len(sig_src)):
        _mean += sig_src[item]
     #total divided by length of list   
    _mean = _mean/len(sig_src)
    return _mean
```

```python
calc_signal_mean(data.InputSignal_1kHz_15kHz)
```

#### Computing the Variance

```python
#variance of the signal
sample_signal_var = np.var(data.InputSignal_1kHz_15kHz)
sample_signal_var
```

```python
_mean = 0.0
_variance = 0.0
def calc_signal_variance(sig_src):
    global _mean
    global _variance
    
     #for each item in the list, add item to total
    for item in range(len(sig_src)):
        _mean += sig_src[item]
     #total divided by length of list   
    _mean = _mean/len(sig_src)
    
    #for each item in the list, subtract mean from signal item 
    #and raise to power of 2.
    for item in range(len(sig_src)):
        _variance = _variance + (sig_src[item] - _mean)**2 
    #total var is divided by length of signal list
    _variance = _variance/len(sig_src)
    return _variance
```

```python
calc_signal_variance(data.InputSignal_1kHz_15kHz)
```

#### Computing the Standard Deviation

```python
sample_signal_std_dev = np.std(data.InputSignal_1kHz_15kHz)
sample_signal_std_dev
```

```python
_mean = 0.0
_variance = 0.0
_std = 0.0
def calc_signal_std_dev(sig_src):
    global _mean
    global _variance
    global _std
    
    #for each item in the list, add item to total
    for item in range(len(sig_src)):
        _mean += sig_src[item]
     #total divided by length of list   
    _mean = _mean/len(sig_src)
    
    #for each item in the list, subtract mean from signal item 
    #and raise to power of 2.
    for item in range(len(sig_src)):
        _variance = _variance + (sig_src[item] - _mean)**2 
    #total var is divided by length of signal list
    _variance = _variance/len(sig_src)
   
    #raise to power of 1/2 is same as sqr root
    _std = _variance**(0.5)
    
    return _std
```

```python
calc_signal_std_dev(data.InputSignal_1kHz_15kHz)
```

#### Signal to Noise Ratio
$$\frac{\mu}{\sigma}$$
- expressed as mean over std dev
- Whilst coefficient of variation is 
$${\frac{\mu}{\sigma}}\cdot100$$

```python
np.mean(data.InputSignal_1kHz_15kHz)/np.std(data.InputSignal_1kHz_15kHz)
```

```python
_mean = 0.0
_variance = 0.0
_std = 0.0
_stn = 0.0
def calc_signal_stn_ratio(sig_src):
    global _mean
    global _variance
    global _std
    global _stn
    
    #for each item in the list, add item to total
    for item in range(len(sig_src)):
        _mean += sig_src[item]
     #total divided by length of list   
    _mean = _mean/len(sig_src)
    
    #for each item in the list, subtract mean from signal item 
    #and raise to power of 2.
    for item in range(len(sig_src)):
        _variance = _variance + (sig_src[item] - _mean)**2 
    #total var is divided by length of signal list
    _variance = _variance/len(sig_src)
   
    #raise to power of 1/2 is same as sqr root
    _std = _variance**(0.5)
    #signal to noise is mean over std dev
    _stn = _mean/_std
    return _stn
```

```python
calc_signal_stn_ratio(data.InputSignal_1kHz_15kHz)
```

```python
def calc_signal_coeff_var(sig_src):
    global _mean
    global _variance
    global _std
    global _coeffvar
    
    #for each item in the list, add item to total
    for item in range(len(sig_src)):
        _mean += sig_src[item]
     #total divided by length of list   
    _mean = _mean/len(sig_src)
    
    #for each item in the list, subtract mean from signal item 
    #and raise to power of 2.
    for item in range(len(sig_src)):
        _variance = _variance + (sig_src[item] - _mean)**2 
    #total var is divided by length of signal list
    _variance = _variance/len(sig_src)
   
    #raise to power of 1/2 is same as sqr root
    _std = _variance**(0.5)
    #coeff of variation is mean over std dev multiplied by 100
    _coeffvar = (_mean/_std)*100
    return _coeffvar
```

```python
calc_signal_coeff_var(data.InputSignal_1kHz_15kHz)
```
